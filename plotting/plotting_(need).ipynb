### Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             color='green', label='Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend()
plt.show()
```

###  We need another axis!

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right') #different legends for different plots 

plt.show()
```

### Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)#export plots into many formats like pdf, jpeg, png, bmp at different resolutions

#when you run this code, you will find the file saved in the same directory (folder) as the one that your notebook lives in. Otherwise youll have to specify the path.
```

## Exercise 2 :  A simple mathematical plot

```python
import numpy as np
from matplotlib import pyplot as plt
x=np.linspace(0,50,1000)
y= np.e**(-x/10)*np.sin(x)
plt.plot(x,y)
plt.grid(alpha=0.25)
plt.xlabel('x',fontsize=15,color="darkblue")
plt.ylabel('y',rotation=0, fontsize=15,color='darkblue')
plt.title('$y=e^{-x/10}\sin(x)$',fontsize=20)
plt.ylim([-1,1])
plt.show()
```
